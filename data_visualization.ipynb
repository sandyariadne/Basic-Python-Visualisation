```python
import pandas as pd
```

```python
df = pd.DataFrame(
[
    ["surabaya", 120, 60],
    ["surabaya", 200, 20],
    ["surabaya", 220, 40],
    ["jakarta", 290, 30],
    ["jogja", 290, 30],
    ["solo", 190, 20],
    ["padang", 170, 40],
    ["lampung", 230, 78],
], columns = ["kota", "angka1", "angka2"]
)
```

```python
import matplotlib.pyplot as plt
```

```python

#
fig, ax = plt.subplots()

#
ax.scatter(df['angka1'], df['angka2'])
#
ax.set_title('Scatter Matplotlib')
ax.set_xlabel('angka1')
ax.set_ylabel('angka2')
```

```python
#create color dictionary
colors = {
    'surabaya':'red',
    'jakarta':'green',
    'jogja':'blue',
    'solo':'yellow',
    'padang':'black',
    'lampung':'purple',
}
#create a figure and axis
fig, ax = plt.subplots()

#scatter the angka1 against the angka2

#plot each data-point
for i in range(len(df['angka1'])):
    ax.scatter(df['angka1'][i],
               df['angka2'][i],
               color=colors[df['kota'][i]])
#set a title and labels
ax.set_title('Colored Scatter plot')
ax.set_xlabel('angka1')
ax.set_ylabel('angka2')
ax.legend(df["kota"])
```

```python
#get columns to plot
columns=df.columns.drop(['kota'])
#create x data
x_data= range(0, df.shape[0])
#create figure and axis
fig, ax = plt.subplots()
#plot each columns
for columns in columns:
    ax.plot(x_data, df[columns])
#set title and legend
ax.set_title('Line chart')
ax.legend(["angka1", "angka2"])
```

```python
#get columns to plot
columns=df.columns.drop(['kota'])
#create figure and axis
fig, ax = plt.subplots()
#plot each columns
for columns in columns:
    ax.plot(df["kota"], df[columns])
#set title and legend
ax.set_title('Line chart')
ax.legend(["angka1", "angka2"])
```

```python
#create a figure and axis
fig, ax = plt.subplots()
#count the occurrence of each class
data = df['kota'].value_counts()
#get x and y data
points = data.index
frequency = data.values
#create bar chart
ax.bar(points, frequency)
#set title and labels
ax.set_title('Bar chart')
ax.set_xlabel('kota')
ax.set_ylabel('frequency')
```
